```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
sonar_data = pd.read_csv('/content/sonar data.csv', header=None)
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].unique()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
X = sonar_data.drop(columns=60, axis=1)
y = sonar_data[60]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)
```

```python
print(X_train.shape,X_test.shape)
```

```python

from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
```

```python




clf.fit(X_train,y_train)
```

```python
y_pred_train = clf.predict(X_train)
accuracy_on_training =accuracy_score(y_train, y_pred_train)
print('accuracy on training data: ',accuracy_on_training)
```

```python
y_pred_test = clf.predict(X_test)
accuracy_on_test =accuracy_score(y_test, y_pred_test)
print('accuracy on test data: ',accuracy_on_test)
```
